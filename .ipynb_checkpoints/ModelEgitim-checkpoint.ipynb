```python
import tensorflow as tf
from tensorflow.keras import layers
from tensorflow.keras.layers import Dense
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from tensorflow.keras.models import load_model
from keras.utils import plot_model
import pickle
```

```python
print(tf.__version__)
```

```python
IMAGE_SIZE = 256
BATCH_SIZE = 32
```

```python
training_dataset = tf.keras.preprocessing.image_dataset_from_directory(
   'Apple_Plant_Disease_Dataset/train',
    shuffle=True,
    image_size = (IMAGE_SIZE,IMAGE_SIZE),
    batch_size = BATCH_SIZE
)

testing_dataset = tf.keras.preprocessing.image_dataset_from_directory(
   'Apple_Plant_Disease_Dataset/test',
    shuffle=True,
    image_size = (IMAGE_SIZE,IMAGE_SIZE),
    batch_size = BATCH_SIZE
)

validation_dataset = tf.keras.preprocessing.image_dataset_from_directory(
   'Apple_Plant_Disease_Dataset/validation',
    shuffle=True,
    image_size = (IMAGE_SIZE,IMAGE_SIZE),
    batch_size = BATCH_SIZE
)
```

```python
class_names = training_dataset.class_names
class_names
```

```python
class_names = testing_dataset.class_names
class_names
```

```python
class_names = validation_dataset.class_names
class_names
```

```python
plt.figure(figsize=(10,10))
for image_batch,label_batch in training_dataset.take(1):
    for i in range(0,8):
        plt.subplot(3,4,i+1)
        plt.imshow(image_batch[i].numpy().astype("uint32"))
        plt.title(class_names[label_batch[i]])
        plt.axis("off")
```

```python
train_data = training_dataset.cache().shuffle(1000).prefetch(buffer_size=tf.data.AUTOTUNE)
valid_data = validation_dataset.cache().shuffle(1000).prefetch(buffer_size=tf.data.AUTOTUNE)
test_data = testing_dataset.cache().shuffle(1000).prefetch(buffer_size=tf.data.AUTOTUNE)
```

```python
resize_scale = tf.keras.Sequential([
    layers.experimental.preprocessing.Resizing(IMAGE_SIZE,IMAGE_SIZE),
    layers.experimental.preprocessing.Rescaling(1.0/255)
])
```

```python
data_augmentation = tf.keras.Sequential([
    layers.experimental.preprocessing.RandomFlip("horizontal_and_vertical"),
    layers.experimental.preprocessing.RandomRotation(0.2)
])
```

```python
input_shape = (32, IMAGE_SIZE, IMAGE_SIZE, 3)

model = tf.keras.Sequential([
    resize_scale,
    data_augmentation,
    layers.Conv2D(filters=32, kernel_size=(3, 3), activation='relu', input_shape=input_shape),
    layers.MaxPool2D((2, 2)),
    layers.Conv2D(64, (3, 3), activation='relu'),
    layers.MaxPool2D((2, 2)),
    layers.Conv2D(64, (3, 3), activation='relu'),
    layers.MaxPool2D((2, 2)),
    layers.Conv2D(64, (3, 3), activation='relu'),  # New Conv2D layer
    layers.MaxPool2D((2, 2)),  # Adding max pooling after the new Conv2D layer
    layers.Conv2D(64, (3, 3), activation='relu'),  # New Conv2D layer
    layers.MaxPool2D((2, 2)),  # Adding max pooling after the new Conv2D layer
    layers.Conv2D(64, (3, 3), activation='relu'),  # New Conv2D layer
    layers.MaxPool2D((2, 2)),  # Adding max pooling after the new Conv2D layer
    layers.Flatten(),
    layers.Dense(64, activation='relu'),
    layers.Dropout(0.5),  # Adding dropout before the output layer (adjust dropout rate as needed)
    layers.Dense(len(class_names), activation='softmax')  # Adjust this value based on your number of classes
])
model.build(input_shape=input_shape)
```

```python
model.compile(
    optimizer='adam',
    loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=False),
    metrics=['accuracy']
)
```

```python
tf.random.set_seed(1)
```

```python
np.random.seed(1)
```

```python
history=model.fit(
    train_data,
    epochs=50,
    batch_size=32,
    verbose=1,
    validation_data=valid_data
)
```

```python
with open('history_dict_96.55', 'wb') as file:
    pickle.dump(history.history, file)
```

```python
# Getting the model history to analyse
train_loss = history['loss']
train_acc = history['accuracy']

val_loss = history['val_loss']
val_acc = history['val_accuracy'] 

# Grafik boyutunu ayarla
plt.figure(figsize=(15, 7))

# Daha fazla veri noktas olutur
epochs = np.linspace(0, 49, num=1000)

# Eitim ve dorulama verilerini interpolasyon yaparak daha przsz hale getir
train_acc_smooth = np.interp(epochs, range(50), train_acc)
val_acc_smooth = np.interp(epochs, range(50), val_acc)
train_loss_smooth = np.interp(epochs, range(50), train_loss)
val_loss_smooth = np.interp(epochs, range(50), val_loss)

# lk alt grafik: Doruluk (Accuracy)
plt.subplot(2, 1, 1)
plt.plot(epochs, train_acc_smooth, label='Training Accuracy', linewidth=2)
plt.plot(epochs, val_acc_smooth, label='Validation Accuracy', linewidth=2)
plt.legend(loc='lower right', fontsize=12)
plt.title('Training and Validation Accuracy', fontsize=18)
plt.xlabel('Epoch', fontsize=16)
plt.ylabel('Accuracy', fontsize=16)
plt.tick_params(axis='both', which='major', labelsize=14)

# kinci alt grafik: Kayp (Loss)
plt.subplot(2, 1, 2)
plt.plot(epochs, train_loss_smooth, label='Training Loss', linewidth=2)
plt.plot(epochs, val_loss_smooth, label='Validation Loss', linewidth=2)
plt.legend(loc='upper right', fontsize=12)
plt.title('Training and Validation Loss', fontsize=18)
plt.xlabel('Epoch', fontsize=16)
plt.ylabel('Loss', fontsize=16)
plt.tick_params(axis='both', which='major', labelsize=14)

# Grafikleri dzenle
plt.tight_layout()

plt.savefig('training_validation_metrics3.png', dpi=1200)
plt.show() 
```

```python
from sklearn.metrics import classification_report

# Test veri kmesi zerinde tahminlerin alnmas
predictions = []
labels = []

for image_batch, label_batch in test_data:
    predictions_batch = model.predict(image_batch)
    predictions.extend(np.argmax(predictions_batch, axis=1))
    labels.extend(label_batch.numpy())

# classification_report ile rapor oluturulmas
report = classification_report(labels, predictions, target_names=class_names)

# Raporun yazdrlmas
print(report)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix
import seaborn as sns

# Test veri kmesi zerinde tahminlerin alnmas
predictions = []
labels = []

for image_batch, label_batch in test_data:
    predictions_batch = model.predict(image_batch)
    predictions.extend(np.argmax(predictions_batch, axis=1))
    labels.extend(label_batch.numpy())

# Karklk matrisinin hesaplanmas
conf_matrix = confusion_matrix(labels, predictions)


# Karklk matrisinin grselletirilmesi
plt.figure(figsize=(8, 4))  # Boyutlar kltld
sns.heatmap(conf_matrix, annot=True, fmt='g', cmap='Greens', xticklabels=class_names, yticklabels=class_names,
            annot_kws={"fontsize": 16}, cbar=True)  

# Eksen etiketleri ve balk font boyutlarn ayarlama
plt.xticks(fontsize=12, rotation=0)  # Eksen etiketlerinin yatay olarak yazdrlmas
plt.yticks(fontsize=12, rotation=0)
plt.xlabel('Predicted', fontsize=12, weight='bold')
plt.ylabel('Actual', fontsize=12, weight='bold')
plt.title('Confusion Matrix', fontsize=12, weight='bold')  # Baln kaln yazlmas

# Karklk matrisini yksek znrlkl olarak kaydetme
plt.savefig('confusion_matrix.png', dpi=1200, bbox_inches='tight')  # dpi deeri artrlarak daha yksek znrlk elde edilir

plt.show()
```

```python
import matplotlib.pyplot as plt

# test_data'tan 3x3'lk bir batch al
for image_batch, label_batch in test_data.take(1):
    # 3x3'lk grid oluturacak ekilde dzenle
    image_batch = image_batch[:9]
    label_batch = label_batch[:9]
    
    # grid'i olutur
    fig = plt.figure(figsize=(10, 10))
    for i in range(9):
        plt.subplot(3, 3, i + 1)
        plt.imshow(image_batch[i].numpy().astype("uint8"), cmap='gray')  # 'gray' renk haritasn kullan
        plt.title("Actual: {}\nPredicted: {}".format(class_names[label_batch[i]], class_names[np.argmax(model.predict(image_batch[i][tf.newaxis]))]))
        plt.axis('off')
        
    fig.tight_layout()
    plt.subplots_adjust(left=0.1, right=0.9, top=0.9, bottom=0.1)
    plt.margins(0.015, tight=True)
        
    plt.savefig('grid_plot1.png', dpi=1200,bbox_inches='tight')
        
    plt.show()
```
